```python
import cv2

def draw_circle(event,x,y,flags,param):
    
    global center, clicked
    
    #track mouse click on down and track center
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        center =  (x,y)
        clicked = False
        
    if event == cv2.EVENT_LBUTTONUP:
        clicked ==  True
        
        #use boolean to  track if the mouse is released
        
#havent drawn anything yet

center = (0,0)
clicked = False

#capture video

cap = cv2.VideoCapture(0)

#create named window for connections
cv2.namedWindow('Test')

#draw circle to the mouse clicks

cv2.setMouseCallback('Test',draw_circle)

while True:
    
    #capture frame by frame
    
    ret,frame = cap.read()
    
    #use if statement if clicked is true
    
    if clicked:
        
        cv2.circle(frame, center = center,radius =50,color = (255,0,0) , thickness = 5)
        
        #display resulting frame
        
    cv2.imshow('Test' , frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
#when everything is done capture

cap.release()

cv2.destroyAllWindows()
    
    
```
