```python
import cv2
import  numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text ='SAIram',org =(50,300),fontFace = font,fontScale = 5,color =(255,255,255),thickness =5)
    return blank_img
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
show_pic(img)
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
```

```python
result = cv2.erode(img,kernel,iterations=1)
```

```python
show_pic(result)
```
